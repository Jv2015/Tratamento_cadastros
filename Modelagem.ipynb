```python
base_columns = ['Id', 'Gender' ,'verified_mobile', 'Document_valid', 'Estado_Civil','VerifiedEmail' , 'Situacao_Cadastral' ]
df_demografico = df[base_columns]

def resp(x):
    if x == True:
        return 'Sim'
    elif x == False:
        return 'Não'

def estadocivil(x):    #Função para corrigir os valores de Estado civil
    if x == 'SINGLE':
        return 'Solteiro'
    elif x == 'SINGLES':
        return 'Solteiro'
    elif x == 'MARRIED':
        return 'Casado'
    elif x == 'AMASSED':
        return 'União Estável'
    elif x == 'AMASSAED ':
        return 'União Estável'
    elif x == 'DIVORCED':
        return 'Divorciado'
    else:
        return 'Viuvo'

def genero(x):
    if x == 'FEMALE':
      return 'Mulher'
    else:
      return 'Homem'

df_demografico.loc[:, 'Document_valid'] = df_demografico['Document_valid'].apply(resp)   #Aplicando as funções
df_demografico.loc[:, 'verified_mobile'] = df_demografico['verified_mobile'].apply(resp)
df_demografico.loc[:, 'Estado_Civil'] = df_demografico['Estado_Civil'].apply(estadocivil)
df_demografico.loc[:, 'VerifiedEmail'] = df_demografico['VerifiedEmail'].apply(resp)
df_demografico.loc[:, 'Gender'] = df_demografico['Gender'].apply(genero)


df_demografico
```

```python
#Criando um dataframe para os endereços informados
base_home = ['Id']
second_home = [x for x in df.columns if x.startswith('home')]   #Primeiro crio um somente com os endereços residenciais
df_home = df[base_home + second_home].copy()
df_home.loc[:, 'Tipo'] = "Residencial"
df_home.rename(columns={'home_address_country': 'País', 'home_address_state': 'Estado', 'home_address_city': 'Cidade'}, inplace=True)

base_comercial = ['Id']
second_comercial = [x for x in df if x.startswith('commercial')]  #Segundo crio um somente com os endereços comerciais
df_comercial = df[base_comercial+second_comercial].copy()
df_comercial.loc[:, 'Tipo'] = "Comercial"
df_comercial.rename(columns={'commercial_address_country': 'País',
                              'commercial_address_state' :'Estado', 'commercial_address_city' : 'Cidade'}, inplace= True)

dict_estados = {'AC': 'Acre', 'AM': 'Amazonas', 'AL': 'Alagoas' , 'AP': 'Amapá', 'BA': 'Bahia', 'CE': 'Ceará', 'DF': 'Distrito Federal',
                'ES' : 'Espírito Santos', 'GO': 'Goiás', 'MA': 'Maranhão', 'MG': 'Minas Gerais', 'MS' :'Mato Grosso do Sul', 'MT': 'Mato Grosso', 'PA': 'Pará',
                'PB': 'Paraíba', 'PE':'Pernambuco'  , 'PI': 'Piauí' , 'PR': 'Paraná', 'RJ' : 'Rio de Janeiro' , 'RN' : 'Rio Grande do Norte',
                'RO' : 'Rondônia', 'RR': 'Rorâima', 'RS' : 'Rio Grande do Sul', 'SC': 'Santa Catarina', 'SE': 'Sergipe', 'SP' : 'São Paulo' , 'TO': 'Tocantins'}

df_enderecos = pd.concat([df_home,df_comercial]).dropna()  #Ao final concateno as duas bases
df_enderecos.Estado = df_enderecos.Estado.map(dict_estados) #Aplico o dicionário criado
df_enderecos
```

```python
#Corrigir porque tem rendas que não batem com que foi informado de patrimonio
base_collumn = ['Id', 'Situacao_Cadastral', 'assets_range', 'income']
df_cadastro_renda = df[base_collumn]

df_cadastro_renda['income'] = pd.to_numeric(df_cadastro_renda['income'], errors='coerce')  # Converter para numérico, forçando NA para valores inválidos
df_cadastro_renda['income'].fillna(0, inplace=True)  # Substituir NA por 0 ou outro valor apropriado
df_cadastro_renda['income'] = df_cadastro_renda['income'].astype(int)  # Converter para int
def renda(x):
  if x in range(0,26000):
    return '0-25 mil'
  elif x in range(26000,51000):
    return '26 -50 mil'
  elif x in range(51000,101000):
    return '51-100 mil'
  elif x in range(101000,201000):
    return '101-200 mil'
  elif x in range(201000,501000):
    return '201-500 mil'
  elif x in range(501000,1100000):
    return '501 mil -1 milhão'
  elif x in range(1100000,10000000):
    return '1,1-10 milhões'
  else:
    return '10-100 milhões'


df_cadastro_renda['Renda'] = df_cadastro_renda.income.apply(renda)
df_cadastro_renda
```

```python
from datetime import datetime
base_collumn = ['Id' , 'Data_Nascimento']
df_idade = df[base_collumn]
df_idade['Data_Nascimento'] = df_idade['Data_Nascimento'].replace('1083-08-27', '27/08/1983') # tive que corrigir na mão esse valor
df_idade['Data_Nascimento'] = df_idade['Data_Nascimento'].replace("0001-01-01", '01/01/2000') # tive que corrigir na mão esse valor
df_idade['Data_Nascimento'] = pd.to_datetime(df_idade['Data_Nascimento'], format='%d/%m/%Y')  #  Padronizei as datas
data_ref = pd.Timestamp('2024-06-14') #data de quando montei
df_idade['Idade'] = df_idade['Data_Nascimento'].apply(lambda x: data_ref.year - x.year - ((data_ref.month, data_ref.day) < (x.month, x.day))).astype(int)
def faixa(x):
  if x in range(0,19):
    return "Até 18 anos"
  elif x in range(19,31):
    return '19 a 30 anos'
  elif x in range(31,46):
    return '30 a 45 anos'
  else:
    return 'Acima de 45 anos'
df_idade['Faixa etária'] = df_idade.Idade.apply(faixa)
df_idade
```
